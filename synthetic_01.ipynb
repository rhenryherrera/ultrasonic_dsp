# This code generates the wavelet and the reflectivity

This code reproduces the synthetic example of the paper: 
Improving Time Estimation by Blind Deconvolution: with Applications to TOFD and Backscatter Sizing. 
Available from: http://www.ndt.net/events/NDTCanada2015/app/content/Paper/8_Herrera_Rev1.pdf


Henry Herrera. henryherrera@ieee.org


# Initialization

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

## Generating the ultrasonic pulse (wavelet)

```python
wn = 0.69;
h = [0.0] * 49
```

```python
for k in range(0,48):
    h[k] = np.cos(wn*(k-22))*np.exp((-(k-24)**2)/100);
```

```python
ax1 = plt.subplot(211)
plt.plot(h)
plt.title('Wavelet - Ultrasound pulse')
```

## Generating the reflectivity

```python
#deltas=[np.zeros((1,49)) 1 zeros(1,399) -0.3*1 zeros(1,49) 0.5*1 zeros(1,300) 0.4 zeros(1,19) -0.9 zeros(1,1320)];
deltas= [0.0] * 1024
deltas[49] = 1.0; deltas[450] = -0.3; deltas[500] = 0.5; deltas[800] = 0.4; deltas[820] = -0.9
```

```python
plt.plot(deltas)
plt.title('Ideal reflectivity')
plt.axis([0.0, 1024.0, -1.0, 1.0])
```

## Convolving reflectivity with the wavelet

```python
y = np.convolve(deltas,h, 'same')
```

```python
plt.plot(y)
plt.title('Synthetic A-scan')
plt.axis([0.0, 1024.0, -1.0, 1.0])
```

## Add noise

```python

noise = 0.08*np.random.randn(1, 1024) ;
yn = y + noise
```

```python
plt.plot(yn[0,:])
plt.title('Noisy Synthetic A-scan')
plt.axis([0.0, 1024.0, -1.0, 1.0])
```
